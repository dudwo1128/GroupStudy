```python
# Diabetes Data 호출

import pandas as pd
import matplotlib.pyplot as plt

data = pd.read_csv('C:/Users/dudwo/Desktop/ML/Data/diabetes.csv')
data.head()
```

```python
# 데이터 분석

data.describe()
```

```python
# Train, Test 데이터 설정

from sklearn.model_selection import train_test_split
X = data.drop(['Outcome','Index'],axis=1)
y = data['Outcome']

X_train,X_test,y_train,y_test = train_test_split(X,y,test_size=0.2)
```

```python
# 데이터 정규화

from sklearn.preprocessing import StandardScaler
sc = StandardScaler()
sc.fit(X_train)

X_train = sc.transform(X_train)
X_test = sc.transform(X_test)
```

```python
# Multilayer Layer Perceptron 적용

from sklearn.neural_network import MLPClassifier
mlp = MLPClassifier(hidden_layer_sizes=(30,30,30),max_iter=100)
mlp.fit(X_train,y_train)

predictions = mlp.predict(X_test)
```

```python
# 모델 평가

from sklearn.model_selection import cross_val_score
scoring = 'Accurcay'
score = cross_val_score(mlp,X_train,y_train,n_jobs=1,cv=5) # cv = cross_validation
print(score)
```
