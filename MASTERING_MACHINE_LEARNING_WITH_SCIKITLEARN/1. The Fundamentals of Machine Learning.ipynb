## Machine Learning
- Learn from data
- Goal : Generalize
- Example : Spam Filtering
- Arthur Samuel : "Machine Learning is the study that gives computers the ability to learn without being explicitly programmed."

## Learning from experience
### Supervised learning
- Learning from pairs of labled inputs and outputs
- Right answers
- Learns from labeled examples of outputs

### Unsuperviesd learning
- Discover patterns

### Semi-supervised learning
- Make use of both supervised and unsupervised data
- Example : Reinforcement Learning

### Keywords
- Output = Response variable
- Input = Features

- Training set = Collection of examples that comprise superviesd experience
- Test set = Collection of examples that is used to assess the performance

## Machine Learning Tasks
### Supervised Machine Learning
1. Classification

    Learn to predict discrete values


2. Regression

    Predict the value of a continuous response variable

### Unsupervised Machine Learning
1. Cluster

    Assigns observation to groups such that observations within groups are more similar to each other


2. Dimensionality reduction

    Process of discovering the explanatory variables that account for the greatest changes in the response variable

    Thousands or even millions of explanatory variables

## Training data and test data
- The observations in the training set comprise the experience that the algorithm uses to learn.

### Test set
- To evaluate the performance
- No observations from the traning set are included in the test set

### Over-fitting
- Memorizing the traning set

### Validation
- A third set of observations
- To tune variables called hyperparameters

### Common Allocation
- 50% : Training set
- 25% : Test set
- Remainder : Validation set

### Cross-Validation
- Train and validate an algorithm on the same data
- Provide more accurate estimate of the model's performance




### Cross- Validation Example
- Partitioned in 5 subsets of equal size
- Labled A,B,C,D,E

## Performance Measures, Bias, and Variance
## Supervised Learning
### Bias, Variance - Fundamental causes of prediction error
#### High Bias
- Produce similar errors for an input
- Under-fits training data


#### High Variance
- Models the noise in the training data
- Over-fits the training data



### Bias-Variance trade-off
- Efforts to decrease one will increase the other

## Unsupervised Learning
- Do not have an 'error signal' to measure
- Performance metrics - measure some attributes of the structure discovered in the data
